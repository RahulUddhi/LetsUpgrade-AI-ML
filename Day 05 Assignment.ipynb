#### first 20 non-even prime natural numbers

```python
num = int(input("Enter a number: "))  

if num > 1: 
    for i in range(2,num):  
        if (num % i) == 0: 
            print(num,"is not a prime number")  
            print(i,"times",num//i,"is",num)  
            break
    else:  
        print(num,"is a prime number")

else:
    print(num,"is not a prime number")
```

# Question 3:
#### palindrome or anagram

```python
def isPalindrome(s): 
    return s == s[::-1] 

def isanagram(s1, s2):
    
    if(sorted(s1)== sorted(s2)): 
        print("The strings are anagrams.")  
    else: 
        print("The strings aren't anagrams.")
s = "malayalam"
ans = isPalindrome(s)
print("Is Palindrome", ans)

s1 ="listen"
s2 ="silent"
ans = isanagram(s1, s2)
```

# Question 4:

#### Remove characters and print in lower case

```python
bad_chars = [';', ':', '!', "*", ',', '.']

def remv_char(test_str): 

    for i in bad_chars : 
        test_str = test_str.replace(i, '')  
    print ("Resultant list is : " + str(test_str.lower()))
    

test_string = '''Python can be easy, to pick up whether you're a first time programmer or you're experienced with other languages!. The following pages are a useful: first 
step to get on your way writing programs with Python!'''
print ("Original String : " + test_string)
test_string = remv_char(test_string)
```
